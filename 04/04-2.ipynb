```python
file = open('04.input.txt', 'r')
lines = file.readlines()

count = 0
for line in lines:
  count += 1

print(count)
```

Helper Line Parser


```python
def parseLine(line, delimiter):
  return line.strip().split(delimiter)
```

List of Drawings

```python
drawings = parseLine(lines[0], ',')

print(drawings)
```

List of Boards

```python
boardsRaw = []

for line in lines[1:]:
  if line == '\n':
    boardsRaw.append([])
  else:
    boardsRaw[-1].append(parseLine(line, None))
```

```python
import numpy as np
boards = []

for board in boardsRaw:
  boards.append(np.array(board))
```

```python
drawnNumbers = []
openBoards = boards.copy()

def checkBoard(board, numbers):
  # check rows
  for row in board:
    if all(drawnNumber in numbers for drawnNumber in row):
      return True
  # check columns
  for column in board.transpose():
    if all(drawnNumber in numbers for drawnNumber in column):
      return True
  return False

def drawNextNumber(nextNumber):
  drawnNumbers.append(nextNumber)

  winningIndeces = []

  for i, board in enumerate(openBoards):
    if checkBoard(board, drawnNumbers):
      print("Heureka! Bingo! Al hamdudila!")
      winningIndeces.append(i)
  return winningIndeces

result = None
for draw in drawings:
  print(draw)
  winningIndeces = drawNextNumber(draw)
  if len(openBoards) == 1 and len(winningIndeces) == 1:
    result = openBoards[0]
    print("found the last winning board: ")
    print(result)
    break
  
  if len(winningIndeces) > 0:
    for winningIndex in winningIndeces[::-1]:
      del openBoards[winningIndex]
    print("removed {} boards from the list - {} boards left open".format(len(winningIndeces), len(openBoards)))
```

```python
def sumOfAllUnmarkedNumbers(board, markedNumbers):
  sum = 0
  for row in board:
    for cell in row:
      if cell not in markedNumbers:
        sum += int(cell)
  return sum


sumOfWinningBoard = sumOfAllUnmarkedNumbers(result, drawnNumbers)
print(sumOfWinningBoard)
lastDrawnNumber = int(drawnNumbers[-1])

print("last drawn number: {}".format(lastDrawnNumber))

score = lastDrawnNumber * sumOfWinningBoard

print("score: {}".format(score))
```
